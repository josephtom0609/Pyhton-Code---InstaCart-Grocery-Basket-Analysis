## Task 4 - Group By with Aggregation using transform()

```python
df_ords_prods_MERGE['Max_Order']=df_ords_prods_MERGE.groupby(['user_id'])['order_number'].transform(np.max)

# Grouping by 'user_id' and the maximum order for each user is calculated
# Max. orders for each user is shown in a new colum 'max_orders'
```

## Task 6 - Spending Flag

```python
df_ords_prods_MERGE['Average_Spend']=df_ords_prods_MERGE.groupby(['user_id'])['prices'].transform(np.mean)

# Grouping by 'user_id' and calclate the average price on all orders made by each user
# Average money spent for each user is shown in a new colum 'average_price'
```

```python
df_ords_prods_MERGE.loc[df_ords_prods_MERGE['Average_Spend'] < 10, 'Spend_Flag'] = 'Low Spender'

# If average spend is less than 10, then Low Spender
```

```python
df_ords_prods_MERGE.loc[df_ords_prods_MERGE['Average_Spend'] >= 10, 'Spend_Flag'] = 'High Spender'

# If average spend is more than or equal to 10, then High Spender
```

```python
df_ords_prods_MERGE.head(10)
```

```python
df_ords_prods_MERGE.dtypes
```

```python
df_ords_prods_MERGE['Average_Spend'] = df_ords_prods_MERGE['Average_Spend'].astype('float16')

# Changing the datatype of the variable 'Average_Spend' from float64 to float16
```

```python
df_ords_prods_MERGE['prices'] = df_ords_prods_MERGE['prices'].astype('float16')

# Changing the datatype of the variable 'Prices' from float64 to float16
```

```python
df_ords_prods_MERGE['Spend_Flag'].value_counts(dropna=False)

# Checking the count of different types of customer based on their spend
```

## Task 7 - Frequency Flag

```python
df_ords_prods_MERGE['Visit_Frequency']=df_ords_prods_MERGE.groupby(['user_id'])['days_since_prior_order'].transform(np.median)

# Grouping by 'user_id' and calculating the median value of last ordered day for all users 
# Frequency of visits for each user is shown in a new column 'Visit Frequency'
```

```python
df_ords_prods_MERGE.loc[df_ords_prods_MERGE['Visit_Frequency'] > 20, 'Frequency_Flag'] = 'Non-Frequent Customer'

# If visit frequency is more than 20, then Non-Frequent Customer
```

```python
df_ords_prods_MERGE.loc[(df_ords_prods_MERGE['Visit_Frequency'] > 10) & (df_ords_prods_MERGE['Visit_Frequency'] <= 20), 'Frequency_Flag'] = 'Regular Customer'

# If visit frequency is more than 10 and less than or equal to 20, then Regular Customer
```

```python
df_ords_prods_MERGE.loc[df_ords_prods_MERGE['Visit_Frequency'] <= 10, 'Frequency_Flag'] = 'Frequent Customer'

# If visit frequency is less than 10, then Frequent Customer
```

```python
df_ords_prods_MERGE['Frequency_Flag'].value_counts(dropna=False)

# Checking the count of different types of customer based on their visits
```
